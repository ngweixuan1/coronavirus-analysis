```python
import pandas as pd
from pycountry_convert import country_name_to_country_alpha3, map_country_alpha3_to_country_name
```

```python
covid = pd.read_csv("owid-covid-data.csv")
```

```python
covid = covid[covid['date'] == '2020-11-17']
covid
```

```python
flights = pd.read_csv("merged-airlines.csv")
flights['iso_code'] = flights['source_airport_country'].apply(lambda x: country_name_to_country_alpha3(x))
flights['dest_airport_country_code'] = flights['dest_airport_country'].apply(lambda x: country_name_to_country_alpha3(x))
```

```python
flights.head()
```

```python
merged = pd.merge(covid, flights, how='inner', on='iso_code')
merged.head()
```

```python
# merged.groupby('dest_airport_country_code').
low_df = pd.DataFrame(columns=['country', 'risk_factor', 'sources'])
merged['multiplied'] = merged['positive_rate'] * merged['flight_capacity']
merged['multiplied'] = merged['multiplied'].fillna(0)
data = []
for country in merged['dest_airport_country_code'].unique():
    merged_country = merged[merged['dest_airport_country_code'] == country]\
        .sort_values('multiplied')
    if country == 'USA':
        print(merged_country.shape)
    total_sources = merged_country.shape[0]/5
    merged_country1 = merged_country[:int(total_sources + 1)]
    countries = ';'.join(merged_country1.iso_code.tolist())

    if country == 'USA':
        print(merged_country1.shape)
        print(countries)
    total_capacity = sum(merged_country1['flight_capacity'])
    datum = {
        'iso_code': country,
        'country': map_country_alpha3_to_country_name()[country],
        'low_risk_factor': sum(merged_country1['multiplied'])/total_capacity,
        'sources': countries
    }
    data.append(datum)
low_df = pd.DataFrame(data, columns=['iso_code','country', 'low_risk_factor', 'sources'])
```

```python
low_df.sort_values('low_risk_factor', ascending=False)
```

```python
# merged.groupby('dest_airport_country_code').
data = []
for country in merged['dest_airport_country_code'].unique():
    merged_country = merged[merged['dest_airport_country_code'] == country]\
        .sort_values('multiplied', ascending=False)
    countries = ";".join(merged_country.iso_code.tolist())
    total_capacity = sum(merged_country['flight_capacity'])
    datum = {
        'iso_code': country,
        'country': map_country_alpha3_to_country_name()[country],
        'high_risk_factor': sum(merged_country['multiplied'])/total_capacity,
        'sources': countries
    }
    data.append(datum)
high_df = pd.DataFrame(data, columns=['iso_code','country', 'high_risk_factor', 'sources'])
```

```python
high_df.sort_values('high_risk_factor', ascending=False)
```

```python
final_df = pd.merge(high_df, low_df, how='inner', on=['iso_code', 'country'])
final_df = final_df.drop('sources_x', axis=1)
final_df.head()
```

```python
final_df.to_csv('risk_factor.csv', index=False)
```
